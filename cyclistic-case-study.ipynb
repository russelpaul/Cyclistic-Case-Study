<font size="3">**Data Cleaning**</font><br>

<br>

<font size="3">**Additional Findings**</font><br>

After cleaning the data, I was able to observe there are rows that have:<br>
**vi.) same start_station_name and end_station_name WITH SAME OR ALMOST SAME values
     as its start & end coordinates (start_lng ≈ end_lng | start_lat ≈ end_lat):**<br>

>**Note that the bikes are geotracked, there are possibilities that some trips are roundtrip (point A = pointB), thus starting point & end point are the same.**

* a.) **where the ride_length is ridiculously SHORT (under a minute)**<br>
         These data can possibly skew our results since some trips have ride length less 60 seconds. Some trips have duration of 1 second only.

* b.) **where the ride_length is acceptable (above a minute)**<br>
         Possibly the user returned the bike exactly where he/she was able to access it.

For both cases, the **‘distance_traveled_km’** is almost or equal to ZERO (0) since start & end coordinates are approximately the same. That is why using ‘distance_traveled_km’ as a condition in the WHERE clause (WHERE distance_traveled_km ≈ 0 ) during data cleaning IS NOT ADVISABLE since some trips that fall in the second category (acceptable ride_length) meets the condition.<br>
 
So instead, **I decided to use the ride_length as a condition to be used in the WHERE clause**, but since the data type of ‘ride_length’ is TIME, I converted it first to FLOAT type.<br>
<br> 

**vii.) different start_station_name and end_station_name**<br>
*  a.) **where the ride_length is ridiculously SHORT for a long distance trip**<br>
         For this to take place, the user’s bike speed should be FAST enough to cover the long distance, a speed that a usual bicycle cannot achieve.

**For both observations, I decided to remove some rows with ride_length less than a minute. (60 seconds)<br>**
